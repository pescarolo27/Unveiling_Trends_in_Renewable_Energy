# Analysis I
**Train a model.**
- Use the Training Set, which contains all features and the target (`Production (GWh)`), to build and fine-tune your model.
- Explore, clean, & transform the data as needed.


Since the target variable, 'Production (GWh),' has continuous numerical values, this process calls for a regression model. Given how many feature variables there are in the datasets (30), linear regression is probably not the most appropriate form of regression to use in this case. Another form, like random forest regression, may be more appropriate because it is more adept at handling more variables that are also potentially more diverse. In this report, both linear & random forest regression were trialed to determine which may produce a more effective model.

On another note, scikit-learn models do not handle non-numerical data. Therefore, values of such variables need to be converted to numerical values which could be done with one-hot encoding.


### Trialing different regression models.

```python
#Define feature & target data
X = train_df.drop(columns=['Production (GWh)'])
y = train_df['Production (GWh)']

#convert categorical variables ('Country', 'Energy Type') to dummy variables
X_dummy = pd.get_dummies(X, drop_first=True)

#split the data into training & testing (validation) sets (using the main training dataset)
X_train, X_val, y_train, y_val = train_test_split(X_dummy, y, test_size=0.2, random_state=10)
```

```python
#initialize models to trial
models = {'Linear Regression':LinearRegression(), 'Random Forest':RandomForestRegressor(random_state=10)}

#train, evaluate the models
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, preds))
    display("{} RMSE: {}".format(name, rmse))

#display statistics of the energy 'Production' variable to assess the RMSE results
display(train_df['Production (GWh)'].describe().reset_index().T)
```

Based on this information above, the random forest model is performing slightly worse than the linear regression model because it has a larger error value. But more or less, they are equivalent based on this metric.

To interpret, these RMSE terms, compare it to the mean of the target variable--48,507. The relative error of both models, as calculated below, are quite high.
- Linear regression: 28538/48507 ~ 58.8%
- Random forest: 28779/48507 ~ 59.3%

Fortunately, there is room for improvement through hyperparameter tuning. This was done for the random forest model rather than the linear regression model because the former is generally more flexible & more adaptable, which is ideal given how many feature variables are being used in this modeling process. This process was limited due to the computational resources available. Trialing too many hyperparameter values would take the system longer than what is reasonable, so a more efficient, stronger system may produce a better model through more sophisticated hyperparameter tuning.

### Hyperparameter tuning (random forest)

```python
#initialize the model
rf = RandomForestRegressor(random_state=10)

#define the hyperparameter grid
param_grid = {'n_estimators':[100, 200],
             'max_depth':[10,20,None],
             'max_features':['sqrt','log2'],
             'min_samples_split':[2,5],
             'min_samples_leaf':[1,2]}

#build the GridSearchCV
CV = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, scoring='neg_root_mean_squared_error', verbose=1, n_jobs=1)

#fit the grid to the training data
CV.fit(X_train, y_train)
```

```python
#NOTE THAT THIS CELL TAKES ABOUT 5-10 MINUTES TO PROCESS

#find the best model & the corresponding score
display("Best RMSE (negative): {}".format(CV.best_score_))
display("Best parameters: {}".format(CV.best_params_))

#evaluate the parameter grid on the validation test (X_val)
best_rf = CV.best_estimator_
y_pred = best_rf.predict(X_val)
rmse = np.sqrt(mean_squared_error(y_val, y_pred))
display("Validation RMSE with tuned model: {}".format(rmse))
```

With these optimal hyperparameter values, they can be used to train the most optimal model on the full training dataset.

```python
#build the model
final_rf = RandomForestRegressor(max_depth=10, max_features='log2', min_samples_leaf=1, min_samples_split=2, 
                                n_estimators=200, random_state=10)
#fit the model to the training data
final_rf.fit(X_train, y_train)
```

```python
#evaluate the new model
final_preds = final_rf.predict(X_val)
final_rmse = np.sqrt(mean_squared_error(y_val, final_preds))
display("Final RMSE on test set: {}".format(final_rmse))
```

This RMSE is slightly better than what it was for both the random forest & linear regression models prior to hyperparameter tuning. Again, more hyperparameter tuning could be performed but a stronger system to run the python notebook would be necessary. 
Nevertheless, this optimized random forest model can be used to cast predictions on a new dataset regarding energy production.

# Analysis II
**Generate predictions.**
- Use your trained model to make predictions for the **Test Set (20%)**, which has all the features *except* `Production (GWh)`.
- The Test Set also has an `ID` **column**, which uniquely identifies each row.

To standardize the modeling process, the new testing dataset needs to be preprocessed just as the training dataset was.

```python
#Convert variable data types
#BINARY variables (0=False, 1=True)
test_df['Government Policies'] = test_df['Government Policies'].astype('bool')
test_df['Renewable Energy Targets'] = test_df['Renewable Energy Targets'].astype('bool')
test_df['Public-Private Partnerships in Energy'] = test_df['Public-Private Partnerships in Energy'].astype('bool')
test_df['Energy Market Liberalization'] = test_df['Energy Market Liberalization'].astype('bool')

#CATEGORICAL variables
test_df['Energy Type'] = test_df['Energy Type'].astype('category')
test_df['Country'] = test_df['Country'].astype('category')


#save the 'ID' column for the final submission
ids = test_df['ID']

#remove this column before encoding dummy variables
test_df = test_df.drop(columns=['ID'])

#encode dummy variables
test_df_dummy = pd.get_dummies(test_df, drop_first=True)

#align the columns of the testing dataset with those of the training dataset
test_df_dummy = test_df_dummy.reindex(columns=X_train.columns, fill_value=0)
```

```python
#generate predictions
test_preds = final_rf.predict(test_df_dummy)

#create a dataframe to house the predictions
prediction_df = pd.DataFrame({'ID':ids, 'Predicted Production (GWh)':test_preds})

#inspect the predictions
display("PREDICTIONS:")
display(prediction_df.head(4))
display(prediction_df['Predicted Production (GWh)'].describe().reset_index().T)

display("----------------------------------------------------------------------------------------------------------------")
display("ORIGINAL DATA:")
display(train_df['Production (GWh)'].describe().reset_index().T)
```

The predicted values lie within the range of those from the original dataset, but they only span a range of about 20,000 gigwatt hours as opposed to almost 100,000 GWh. The mean & median values of the two distributions are within about 30 GWh & 1,500 GWh respectively. The main difference between these two distributions is the range of values. Note that the predictions only contain a quarter of the number of data points as the original dataset.
